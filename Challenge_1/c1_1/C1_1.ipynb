## Imports

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

# PART I

## Prepare Data

<b>A. Create a data set D with 100 data points with the following attributes (features) as such:</b>

a1 is from a normal distribution N(mean=20,var=4)<br>
a2 is from a uniform distribution U(min=14,max=26)<br>

```python
np.random.seed(0) # Create reproducible random values

# General variables
n = 100

# Generation of a1
a1_mean = 20
a1_var = 4
a1 = np.random.normal(loc=a1_mean, scale=np.sqrt(a1_var), size=n)

# Generation of a2
a2_min = 14
a2_max = 26
a2 = np.random.uniform(low=a2_min, high=a2_max, size=n)

# Generation of data set D
D = pd.DataFrame({
    'a1': a1,
    'a2': a2
})

print("Data Set D \n=========================")
print(D)
```

## Explore Data

Apply min-max scaling to D, P1:

```python
P1 = MinMaxScaler().fit_transform(D)
print(P1)
```

Apply std-mean scaling to D, P2

```python
P2 = StandardScaler().fit_transform(D)
print(P2)
```

Apply min-max scaling to P2, P3:

```python
P3 = MinMaxScaler().fit_transform(P2)
print(P3)
```

Plot P1, P2, and P3 for each attribute on a separate chart. Each sub-plot must be side-by-side:

```python
transformations = {
    'P1 (MinMax)': MinMaxScaler().fit_transform(D),
    'P2 (Standard)': StandardScaler().fit_transform(D),
    'P3 (MinMax of P2)': MinMaxScaler().fit_transform(StandardScaler().fit_transform(D))
}

fig, axes = plt.subplots(2, 3, figsize=(18, 8))

for i, (title, P) in enumerate(transformations.items()):
    df = pd.DataFrame(P, columns=['a1', 'a2']) 
    for j, attr in enumerate(['a1', 'a2']):
        axes[j, i].plot(df[attr], linestyle='-', linewidth=2)
        axes[j, i].set_title(f'{title} - {attr}')
        axes[j, i].set_xlabel('Index')
        axes[j, i].set_ylabel('Transformed Value')
        axes[j, i].grid(True)

plt.tight_layout()
plt.savefig("c1-1_plot1.png")
plt.show()


fig, axes = plt.subplots(2, 3, figsize=(18, 8)) 

for i, (title, P) in enumerate(transformations.items()):
    df = pd.DataFrame(P, columns=['a1', 'a2']) 
    for j, attr in enumerate(['a1', 'a2']):
        axes[j, i].hist(df[attr], bins=15, alpha=0.7, edgecolor='black')
        axes[j, i].set_title(f'{title} - {attr}')
        axes[j, i].set_xlabel('Value')
        axes[j, i].set_ylabel('Frequency')
        axes[j, i].grid(True)

plt.tight_layout()
plt.savefig("c1-1_plot2.png")
plt.show()
```

Compare P1 and P3. Discuss the relation between P1 and P3:

P1 applies a MinMax scaling directly to the original data set. which preserves its distribution while rescaling values between 0 and 1. P3, on the otherhand, starts by standizing the data set P2 to have a mean if 0 and a STD of 1 before applying the MinMax scaling. Meaning that P3 adjusts for the differences in variance, potentially altering the original spread of the values, whereas P2 maintins the original relative distance.
